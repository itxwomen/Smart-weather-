```python
#–°–æ–≤–µ—Ç—ã –ø–æ –ø–æ–≥–æ–¥–µ
while True:
    try: 
        temp = int(input("–°–∫–æ–ª—å–∫–æ —Å–µ–≥–æ–¥–Ω—è –≥—Ä–∞–¥—É—Å–æ–≤?")) #—É–∑–Ω–∞–µ–º –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –ø–æ –ø–æ–≥–æ–¥–µ —É –ø–æ–ª—å–∑–æ–≤–∞—Ç–µ–ª—è
    except ValueError:
        print("–û—à–∏–±–∫–∞! –í–≤–µ–¥–∏—Ç–µ —á–∏—Å–ª–æ")
        continue
    
    rain = input("–ò–¥–µ—Ç –¥–æ–∂–¥—å? (–¥–∞/–Ω–µ—Ç)")
    if rain.lower() not in ["–¥–∞", "–Ω–µ—Ç"]: #–ø—Ä–æ–≤–µ—Ä—è–º –∫–æ—Ä—Ä–µ–∫—Ç–Ω–æ—Å—Ç—å –≤–≤–æ–¥–∞
         print("–û—à–∏–±–∫–∞ –≤–≤–µ–¥–∏—Ç–µ '–¥–∞' –∏–ª–∏ '–Ω–µ—Ç'")
         continue 
    print ("\t–°–û–í–ï–¢–´ –ü–û –ü–û–ì–û–î–ï –ù–ê –°–ï–ì–û–î–ù–Ø:") #–¥–∞–µ–º –Ω—É–∂–Ω—ã–µ —Å–æ–≤–µ—Ç—ã –≤—Å–≤—è–∑–∏ —Å –ø–æ–≥–æ–¥–æ–π
    if temp <=0:
         print ("–†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏ –ø–æ –æ–¥–µ–∂–¥–µ:\n–ó–∏–º–Ω—è—è –∫—É—Ä—Ç–∫–∞, —à–∞—Ä—Ñ, —à–∞–ø–∫–∞, –±–æ—Ç–∏–Ω–∫–∏ –∏ —Ç–µ–ø–ª—ã–µ –ø–æ–¥—à—Ç–∞–Ω–∏–∫–∏\n–°–µ–≥–æ–¥–Ω—è —Ö–æ–ª–æ–¥–Ω–æ, –æ–¥–µ–≤–∞–π—Å—è —Ç–µ–ø–ª–æ")
    elif temp >0 and temp<20:
         print ("–†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏ –ø–æ –æ–¥–µ–∂–¥–µ:\n–û–¥–µ–≤–∞–π –±–∞–π–∫—É, –∫—Ä–æ—Å—Å–æ–≤–∫–∏ –∏ —à—Ç–∞–Ω—ã\n–°–µ–≥–æ–¥–Ω—è –ø—Ä–æ—Ö–ª–∞–¥–Ω–æ, –æ–¥–µ–≤–∞–π—Å—è —Ç–µ–ø–ª–µ–π")
    else:
         print ("–†–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏ –ø–æ –æ–¥–µ–∂–¥–µ:\n–ú–∞–π–∫–∞, –∫—Ä–æ—Å—Å–æ–≤–∫–∏/–±–æ—Å–æ–Ω–æ–∂–∫–∏ –∏ —à–æ—Ä—Ç—ã\n–ù–∞ —É–ª–∏—Ü–µ –∂–∞—Ä–∫–æ, –æ–¥–µ–≤–∞–π—Å—è –ª–µ–≥—á–µ") 
    if   rain.lower() == "–¥–∞": #–ø—Ä–∏–≤–æ–¥–∏–º –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É
         print("- –ü—Ä–∏—Ö–≤–∞—Ç–∏ –∑–æ–Ω—Ç")
    else:
         print ("- –ó–æ–Ω—Ç –Ω–µ –Ω—É–∂–µ–Ω")
    break #–≤—ã—Ö–æ–¥–∏–º –∏–∑ —Ü–∏–∫–ª–∞ –ø–æ—Å–ª–µ —É—Å–ø–µ—à–Ω–Ω–æ–≥–æ –≤—ã–ø–æ–ª–Ω–µ–Ω–∏—è
 
```
